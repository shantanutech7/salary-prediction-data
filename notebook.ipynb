```python
# First, let's re-create a sample DataFrame with some null values for demonstration purposes
import numpy as np
df_sample = df.copy()
df_sample.loc[0, 'YearsExperience'] = np.nan
df_sample.loc[5, 'Salary'] = np.nan
df_sample.loc[10, 'YearsExperience'] = np.nan

print("Sample DataFrame with introduced null values:")
display(df_sample.head(15))
print("\nNull values in the sample DataFrame before filling:")
display(df_sample.isnull().sum())
```

```python
# Code to fill null values:

# Fill null values for numerical columns with the mean
for col in df_sample.select_dtypes(include=np.number).columns:
    if df_sample[col].isnull().any():
        mean_val = df_sample[col].mean()
        # Reassign the column to avoid FutureWarning about inplace=True in chained assignment
        df_sample[col] = df_sample[col].fillna(mean_val)

# For categorical columns, fill with the mode.
# (Note: 'Salary_Data.csv' does not have categorical columns, this is for general use)
for col in df_sample.select_dtypes(include='object').columns:
    if df_sample[col].isnull().any():
        mode_val = df_sample[col].mode()[0]
        # Reassign the column to avoid FutureWarning about inplace=True in chained assignment
        df_sample[col] = df_sample[col].fillna(mode_val)

print("Sample DataFrame after filling null values:")
display(df_sample.head(15))
print("\nNull values in the sample DataFrame after filling:")
display(df_sample.isnull().sum())
```

This code successfully fills the null values in the `df_sample` DataFrame. Remember, your original `df` DataFrame currently does not have any missing values.

### Demonstrating Label Encoding

To demonstrate label encoding, we need a DataFrame with categorical features. Let's create a new sample DataFrame for this purpose.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Create a sample DataFrame with a categorical column
data = {
    'Color': ['Red', 'Green', 'Blue', 'Green', 'Red', 'Yellow', 'Blue'],
    'Size': ['Small', 'Medium', 'Large', 'Medium', 'Small', 'Medium', 'Large'],
    'Price': [10, 20, 30, 25, 15, 35, 40]
}
df_categorical = pd.DataFrame(data)

print("Original DataFrame with categorical data:")
display(df_categorical)
```

```python
# Initialize LabelEncoder
le = LabelEncoder()

# Apply Label Encoding to the 'Color' column
df_categorical['Color_Encoded'] = le.fit_transform(df_categorical['Color'])

print("DataFrame after applying Label Encoding to 'Color' column:")
display(df_categorical)

print("\nMapping of original categories to encoded labels:")
for i, item in enumerate(le.classes_):
    print(f"{item} -> {i}")
```

### Splitting Data into Training and Testing Sets

Now that our data is prepared (or we've demonstrated how to handle preparation), a crucial step in machine learning is to split the data into training and testing sets. This allows us to train a model on one portion of the data and evaluate its performance on an independent, unseen portion.

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
# Assuming 'YearsExperience' is the feature and 'Salary' is the target for a simple model
X = df[['YearsExperience']]
y = df['Salary']

# Split the data into training and testing sets
# test_size=0.20 means 20% of the data will be used for testing
# random_state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)

print("\nFirst 5 rows of X_train:")
display(X_train.head())
print("\nFirst 5 rows of y_train:")
display(y_train.head())
```

We have successfully split the `df` DataFrame into training and testing sets. `X_train` and `y_train` are used for training a machine learning model, while `X_test` and `y_test` are reserved for evaluating the model's performance on new, unseen data. In this split, 80% of the data is for training and 20% is for testing.

### 1. Linear Regression

Linear Regression is a statistical method used for modeling the relationship between a scalar dependent variable `y` and one or more explanatory variables `X`. In our case, we'll use `YearsExperience` to predict `Salary`.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Initialize and train the Linear Regression model
linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_lr = linear_reg_model.predict(X_test)

# Evaluate the model
r2_lr = r2_score(y_test, y_pred_lr)
print(f"Linear Regression R-squared: {r2_lr:.2f}")
```

Now, let's proceed with a **Decision Tree Regressor**. Decision trees are non-parametric supervised learning methods used for both classification and regression. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features.

```python
from sklearn.tree import DecisionTreeRegressor

# Initialize and train the Decision Tree Regressor model
dtree_reg_model = DecisionTreeRegressor(random_state=42)
dtree_reg_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_dt = dtree_reg_model.predict(X_test)

# Evaluate the model
r2_dt = r2_score(y_test, y_pred_dt)
print(f"Decision Tree Regressor R-squared: {r2_dt:.2f}")
```

Next, we'll implement a **Random Forest Regressor**. Random Forests are an ensemble learning method for regression (and classification) that operates by constructing a multitude of decision trees at training time and outputting the mean prediction of the individual trees.

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize and train the Random Forest Regressor model
rf_reg_model = RandomForestRegressor(random_state=42)
rf_reg_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = rf_reg_model.predict(X_test)

# Evaluate the model
r2_rf = r2_score(y_test, y_pred_rf)
print(f"Random Forest Regressor R-squared: {r2_rf:.2f}")
```

Let's also build a **K-Nearest Neighbors (KNN) Regressor**. KNN is a non-parametric, lazy learning algorithm. For regression, the output is the mean of the values of the k-nearest neighbors.

```python
from sklearn.neighbors import KNeighborsRegressor

# Initialize and train the K-Nearest Neighbors Regressor model
knn_reg_model = KNeighborsRegressor()
knn_reg_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_knn = knn_reg_model.predict(X_test)

# Evaluate the model
r2_knn = r2_score(y_test, y_pred_knn)
print(f"K-Nearest Neighbors Regressor R-squared: {r2_knn:.2f}")
```

Regarding **Naive Bayes**, it's primarily a classification algorithm and is generally not directly applicable to regression problems like predicting `Salary`. If you are interested in a classification task based on this data (e.g., classifying salary into high/medium/low categories), we can explore that separately. For now, we'll focus on the regression models.

### Evaluating Regression Models with MAE, MSE, and RMSE

In addition to R-squared, Mean Absolute Error (MAE), Mean Squared Error (MSE), and Root Mean Squared Error (RMSE) are crucial metrics for evaluating regression models. Let's calculate these for each of our models.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# --- Linear Regression Metrics ---
mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mse_lr)

print(f"Linear Regression Metrics:")
print(f"  MAE: {mae_lr:.2f}")
print(f"  MSE: {mse_lr:.2f}")
print(f"  RMSE: {rmse_lr:.2f}")
print(f"  R-squared: {r2_lr:.2f}")
print("\n")

# --- Decision Tree Regressor Metrics ---
mae_dt = mean_absolute_error(y_test, y_pred_dt)
mse_dt = mean_squared_error(y_test, y_pred_dt)
rmse_dt = np.sqrt(mse_dt)

print(f"Decision Tree Regressor Metrics:")
print(f"  MAE: {mae_dt:.2f}")
print(f"  MSE: {mse_dt:.2f}")
print(f"  RMSE: {rmse_dt:.2f}")
print(f"  R-squared: {r2_dt:.2f}")
print("\n")

# --- Random Forest Regressor Metrics ---
mae_rf = mean_absolute_error(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)

print(f"Random Forest Regressor Metrics:")
print(f"  MAE: {mae_rf:.2f}")
print(f"  MSE: {mse_rf:.2f}")
print(f"  RMSE: {rmse_rf:.2f}")
print(f"  R-squared: {r2_rf:.2f}")
print("\n")

# --- K-Nearest Neighbors Regressor Metrics ---
mae_knn = mean_absolute_error(y_test, y_pred_knn)
mse_knn = mean_squared_error(y_test, y_pred_knn)
rmse_knn = np.sqrt(mse_knn)

print(f"K-Nearest Neighbors Regressor Metrics:")
print(f"  MAE: {mae_knn:.2f}")
print(f"  MSE: {mse_knn:.2f}")
print(f"  RMSE: {rmse_knn:.2f}")
print(f"  R-squared: {r2_knn:.2f}")
```

You can then load this model back into your environment like this:

```python
# Load the model from the .pkl file
with open(model_filename, 'rb') as file:
    loaded_linear_reg_model = pickle.load(file)

print(f"Linear Regression model loaded from {model_filename}")

# You can now use the loaded_linear_reg_model for predictions
# For example:
# loaded_y_pred_lr = loaded_linear_reg_model.predict(X_test)
```
